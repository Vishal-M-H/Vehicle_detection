## 2. Train YOLOv8n Model

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

model.train(
    data="/content/dataset/data.yaml",
    epochs=30,
    batch=32,
    imgsz=640,
    device=0,
    save=True,
    save_period=1,   # ⭐ Save weights every epoch
    project="vehicle_detection",
    name="yolov8n_run"
)
```

## 3. Find All Saved Model Checkpoints

```python
import glob

# Find ALL model files
model_paths = glob.glob("/content/runs/detect/**/*.pt", recursive=True)

print("Found models:\n")
for p in model_paths:
    print(p)
```

## 4. Evaluate All Checkpoints on Test Set

```python
from ultralytics import YOLO

data_yaml = "/content/dataset/data.yaml"

results = []

for path in model_paths:
    print(f"\n🔎 Evaluating: {path}")

    model = YOLO(path)
    metrics = model.val(data=data_yaml, split="test")

    map50_95 = metrics.box.map
    map50 = metrics.box.map50

    results.append((path, map50_95, map50))

# Sort by best performance
results.sort(key=lambda x: x[1], reverse=True)
```

## 5. Display Best Model

```python
print("\n🏆 BEST MODEL BASED ON TEST SET:")

best = results[0]

print("Path:", best[0])
print("mAP50-95:", best[1])
print("mAP50:", best[2])
```
